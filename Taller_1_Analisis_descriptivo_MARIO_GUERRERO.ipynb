# Taller 1 - Análisis descriptivo

Para el desarrollo de este notebook, se utilizó una base de datos proveniente de [kaggle](https://www.kaggle.com/datasets/imakash3011/customer-personality-analysis) que contiene información de



* People

  -  ID: Customer's unique identifier
  -  Year_Birth: Customer's birth year
  -  Education: Customer's education level
  -  Marital_Status: Customer's marital status
  -  Income: Customer's yearly household income
  -  Kidhome: Number of children in customer's household
  -  Teenhome: Number of teenagers in customer's household
  -  Dt_Customer: Date of customer's enrollment with the company
  -  Recency: Number of days since customer's last purchase
  -  Complain: 1 if the customer complained in the last 2 years, 0 otherwise

* Products

  - MntWines: Amount spent on wine in last 2 years

  - MntFruits: Amount spent on fruits in last 2 years

  - MntMeatProducts: Amount spent on meat in last 2 years

  - MntFishProducts: Amount spent on fish in last 2 years

  - MntSweetProducts: Amount spent on sweets in last 2 years
  
  - MntGoldProds: Amount spent on gold in last 2 years

* Promotion
  
  - NumDealsPurchases: Number of purchases made with a discount
  
  - AcceptedCmp1: 1 if customer accepted the offer in the 1st campaign, 0 otherwise

  - AcceptedCmp2: 1 if customer accepted the offer in the 2nd campaign, 0 otherwise

  - AcceptedCmp3: 1 if customer accepted the offer in the 3rd campaign, 0 otherwise

  - AcceptedCmp4: 1 if customer accepted the offer in the 4th campaign, 0 otherwise

  - AcceptedCmp5: 1 if customer accepted the offer in the 5th campaign, 0 otherwise

  - Response: 1 if customer accepted the offer in the last campaign, 0 otherwise

* Place
  - NumWebPurchases: Number of purchases made through the company’s website
  - NumCatalogPurchases: Number of purchases made using a catalogue
  - NumStorePurchases: Number of purchases made directly in stores
  - NumWebVisitsMonth: Number of visits to company’s website in the last month

# Visualización de datos


## ``Matplolib``



Gráfico de barras

Realice un gráfico de barras de los clientes por el estado civil

```python
marital_status_counts = df['Marital_Status'].value_counts()

plt.figure(figsize=(10, 6))
marital_status_counts.plot(kind='bar', color='skyblue')
plt.title('Distribución de Clientes por Estado Civil')
plt.xlabel('Estado Civil')
plt.ylabel('Número de Clientes')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

**ANALISIS**
El gráfico muestra que la mayoría de los clientes están casados, seguidos por quienes están en unión (“together”) y solteros, lo que indica que el grueso de la base se concentra en relaciones estables o individuales. Los divorciados y viudos representan una proporción menor, y las categorías “alone”, “absurd” y “YOLO” son prácticamente irrelevantes. En conjunto, la distribución está claramente dominada por los estados civiles tradicionales.

Realice un gráfico de pastel de los clientes por estado civil

```python
plt.figure(figsize=(10, 10))

# Define explode for smaller slices to ensure visibility.
# The `marital_status_counts` series is already sorted by value in descending order.
# Categories are: Married, Together, Single, Divorced, Widow, Alone, Absurd, YOLO
# We will apply a larger explode to the smaller categories.
explode = (0, 0, 0, 0, 0.1, 0.2, 0.3, 0.4)

marital_status_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors,
                           explode=explode, pctdistance=0.85, labeldistance=1.1)
plt.title('Distribución de Clientes por Estado Civil', y=1.05) # Adjusted y-position for the title
plt.ylabel('') # Hide the default 'Marital_Status' y-label
plt.tight_layout()
plt.show()
```

**ANALISIS**
Este gráfico circular refuerza lo visto antes: la mayor proporción de clientes está casada (38,7%), seguida por quienes están en unión “together” (25,9%) y solteros (21,3%). Los divorciados (10,5%) y viudos (3,4%) representan segmentos mucho más pequeños, mientras que categorías como “alone” y “YOLO” son prácticamente insignificantes. En conjunto, más del 85% de los clientes se concentra en los tres estados civiles principales, lo que evidencia una base mayoritariamente tradicional.

Realice un histograma para las personas que tienen ingresos menores a 200000

```python
df_filtered_income = df[df['Income'] < 200000]

plt.figure(figsize=(10, 6))
plt.hist(df_filtered_income['Income'].dropna(), bins=30, color='lightgreen', edgecolor='black')
plt.title('Distribución de Ingresos para Clientes con Ingresos < 200,000')
plt.xlabel('Ingresos')
plt.ylabel('Frecuencia')
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.show()
```

**ANALISIS**
Este histograma muestra que los ingresos de los clientes (menores a 200.000) se concentran principalmente entre 30.000 y 80.000, donde se observa la mayor frecuencia. La distribución es asimétrica a la derecha, con una cola que se extiende hacia ingresos más altos, lo que indica pocos clientes con ingresos significativamente mayores. También se aprecia una menor cantidad de clientes en los niveles de ingreso más bajos, sugiriendo que la base principal corresponde a ingresos medios dentro del rango analizado.

Haga un diagrama de Box-plot para los ingresos

```python
plt.figure(figsize=(8, 6))
plt.boxplot(df['Income'].dropna())
plt.title('Diagrama de Caja de Ingresos')
plt.ylabel('Ingresos')
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

**ANALISIS**
El diagrama de caja muestra que la mayoría de los ingresos se concentra en un rango relativamente estrecho, con una mediana alrededor de valores medios. Se observa asimetría positiva, ya que existen varios valores atípicos altos, incluyendo uno extremadamente elevado, que se aleja claramente del resto de los datos. Esto indica que, aunque la mayoría de los clientes tiene ingresos moderados, hay pocos casos con ingresos muy superiores que influyen en la dispersión total.

Realice un diagrama de dispersión para los clientes que tienen ingresos menores a 200000 versus la cantidad gastada en vino en los ultimos 2 años

```python
df_filtered = df[df['Income'] < 200000]

plt.figure(figsize=(10, 6))
plt.scatter(df_filtered['Income'], df_filtered['MntWines'], alpha=0.7, color='purple')
plt.title('Ingresos vs. Gasto en Vino (Ingresos < 200,000)')
plt.xlabel('Ingresos')
plt.ylabel('Cantidad Gastada en Vino (últimos 2 años)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

**ANALISIS**
Este diagrama de dispersión muestra una relación positiva entre los ingresos y el gasto en vino: a medida que aumentan los ingresos, el gasto tiende a ser mayor. Sin embargo, la dispersión de los puntos indica que la relación no es perfectamente lineal, ya que existen clientes con ingresos similares pero niveles de gasto muy distintos. También se observan valores atípicos, tanto de ingresos altos con bajo gasto como de gastos elevados, lo que sugiere comportamientos de consumo heterogéneos dentro del segmento analizado.
